```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import skewnorm, gamma

# Function to estimate mode from histogram
def estimate_mode(data):
    counts, bins = np.histogram(data, bins=100, density=True)
    bin_centers = (bins[:-1] + bins[1:]) / 2
    mode_index = np.argmax(counts)
    return bin_centers[mode_index]

# -------------------------------
# Right-Skewed (Positive Skew) Distribution - Gamma
# -------------------------------
shape, scale = 2, 2
right_skewed_data = np.random.gamma(shape, scale, size=10000)

mean_right = np.mean(right_skewed_data)
median_right = np.median(right_skewed_data)
mode_right = estimate_mode(right_skewed_data)

# Quartiles
q1_right, q2_right, q3_right = np.percentile(right_skewed_data, [25, 50, 75])

# Plot right-skewed
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
counts, bins, _ = plt.hist(right_skewed_data, bins=100, density=True, alpha=0.7, color='skyblue', edgecolor='black')
plt.title('Right-Skewed (Positive Skew)\n(Gamma Distribution)')
plt.xlabel('Value')
plt.ylabel('Density')

# Plot mean, median, mode
plt.axvline(mean_right, color='red', linestyle='-', linewidth=2, label=f'Mean = {mean_right:.2f}')
plt.axvline(median_right, color='green', linestyle='-', linewidth=2, label=f'Median = {median_right:.2f}')
plt.axvline(mode_right, color='blue', linestyle='-', linewidth=2, label=f'Mode ≈ {mode_right:.2f}')

# Plot quartiles
plt.axvline(q1_right, color='orange', linestyle='--', linewidth=2, label=f'Q1 (25th %ile) = {q1_right:.2f}')
plt.axvline(q2_right, color='green', linestyle='--', linewidth=2, label=f'Q2 (50th %ile) = {q2_right:.2f}')  # Same as median
plt.axvline(q3_right, color='purple', linestyle='--', linewidth=2, label=f'Q3 (75th %ile) = {q3_right:.2f}')

plt.legend(fontsize=9)
plt.grid(True, alpha=0.3)

# -------------------------------
# Left-Skewed (Negative Skew) Distribution - Skew-Normal
# -------------------------------
a = -5
left_skewed_data = skewnorm.rvs(a, size=10000)

mean_left = np.mean(left_skewed_data)
median_left = np.median(left_skewed_data)
mode_left = estimate_mode(left_skewed_data)

# Quartiles
q1_left, q2_left, q3_left = np.percentile(left_skewed_data, [25, 50, 75])

# Plot left-skewed
plt.subplot(1, 2, 2)
counts, bins, _ = plt.hist(left_skewed_data, bins=100, density=True, alpha=0.7, color='salmon', edgecolor='black')
plt.title('Left-Skewed (Negative Skew)\n(Skew-Normal Distribution)')
plt.xlabel('Value')
plt.ylabel('Density')

# Plot mean, median, mode
plt.axvline(mean_left, color='red', linestyle='-', linewidth=2, label=f'Mean = {mean_left:.2f}')
plt.axvline(median_left, color='green', linestyle='-', linewidth=2, label=f'Median = {median_left:.2f}')
plt.axvline(mode_left, color='blue', linestyle='-', linewidth=2, label=f'Mode ≈ {mode_left:.2f}')

# Plot quartiles
plt.axvline(q1_left, color='orange', linestyle='--', linewidth=2, label=f'Q1 (25th %ile) = {q1_left:.2f}')
plt.axvline(q2_left, color='green', linestyle='--', linewidth=2, label=f'Q2 (50th %ile) = {q2_left:.2f}')
plt.axvline(q3_left, color='purple', linestyle='--', linewidth=2, label=f'Q3 (75th %ile) = {q3_left:.2f}')

plt.legend(fontsize=9)
plt.grid(True, alpha=0.3)

# Final layout
plt.tight_layout()
plt.show()

# Print skewness and quartile ranges
print(f"Right-skewed data - Skewness: {stats.skew(right_skewed_data):.3f}")
print(f"Right-skewed data - IQR (Q3 - Q1): {q3_right - q1_right:.3f}")
print(f"Left-skewed data - Skewness: {stats.skew(left_skewed_data):.3f}")
print(f"Left-skewed data - IQR (Q3 - Q1): {q3_left - q1_left:.3f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_quadratic_with_bounds(a0, a1, a2, c1, c2, n0, omega_max=20, num_points=1000):
    """
    Plot a quadratic function f(ω) = a0 + a1*ω + a2*ω²
    and overlay its Big O (upper) and Big Omega (lower) asymptotic bounds.

    Parameters:
    - a0, a1, a2: coefficients of the quadratic function
    - c1: constant for Big O upper bound: U(ω) = c1 * ω²
    - c2: constant for Big Omega lower bound: L(ω) = c2 * ω²
    - n0: threshold where bounds become valid (ω ≥ n0)
    - omega_max: maximum value of ω to plot
    - num_points: number of points for smooth plotting
    """
    
    # Generate ω values
    omega = np.linspace(0, omega_max, num_points)
    
    # Compute f(ω) = a0 + a1*ω + a2*ω²
    f_omega = a0 + a1 * omega + a2 * omega**2
    
    # Compute asymptotic bounds
    upper_bound = c1 * omega**2      # Big O: f(ω) = O(ω²)
    lower_bound = c2 * omega**2      # Big Omega: f(ω) = Ω(ω²)
    
    # Create plot
    plt.figure(figsize=(8, 6))
    
    # Plot the actual function
    plt.plot(omega, f_omega, 'b-', linewidth=2.5, label=r'$f(\omega) = a_0 + a_1\omega + a_2\omega^2$')
    
    # Plot upper bound (Big O)
    plt.plot(omega, upper_bound, 'r--', linewidth=1.5, label=r'$c_1 \omega^2$ (Big O)')
    
    # Plot lower bound (Big Omega)
    plt.plot(omega, lower_bound, 'g--', linewidth=1.5, label=r'$c_2 \omega^2$ (Big $\Omega$)')
    
    # Vertical line at ω = n0
    plt.axvline(x=n0, color='purple', linestyle='-.', linewidth=2, 
                label=f'Verification starts at $\\omega = n_0 = {n0}$')
    
    # Shade region where bounds are valid (ω >= n0)
    omega_shade = np.linspace(n0, omega_max, num_points)
    f_shade = a0 + a1 * omega_shade + a2 * omega_shade**2
    upper_shade = c1 * omega_shade**2
    lower_shade = c2 * omega_shade**2
    
    plt.fill_between(omega_shade, lower_shade, upper_shade, color='yellow', alpha=0.2, 
                     label=r'$c_2 \omega^2 \leq f(\omega) \leq c_1 \omega^2$ for $\omega \geq n_0$')
    
    # Optional: highlight f(ω) within bounds in shaded region
    plt.plot(omega_shade, f_shade, 'b-', linewidth=2.5)
    
    # Labels and formatting
    plt.title(r'Quadratic Function with Big O and Big Omega Asymptotic Bounds', fontsize=16)
    plt.xlabel(r'$\omega$', fontsize=14)
    plt.ylabel(r'$f(\omega)$', fontsize=14)
    plt.legend(loc='upper left', fontsize=11)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    
    # Set limits for better visualization
    plt.xlim(0, omega_max)
    plt.ylim(bottom=-1)  # Avoid negative y if unnecessary
    
    plt.tight_layout()
    plt.show()
```

```python

# ========== EXAMPLE USAGE ==========
# Define your quadratic coefficients
a0 = 1
a1 = 2
a2 = 1

# Define asymptotic bound constants and threshold
c1 = 5   # Upper bound constant (Big O)
c2 = 1/2   # Lower bound constant (Big Omega)
n0 = 2    # Threshold beyond which bounds hold

# Plot!
plot_quadratic_with_bounds(a0, a1, a2, c1, c2, n0, omega_max=200)
```
